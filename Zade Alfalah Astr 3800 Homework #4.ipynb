```python
from math import sqrt
from math import pi
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def binom(n, k, p):
    # defines choosing as x
    x = (math.factorial(n)) / (math.factorial(k) * math.factorial(n - k))
    final = x * p ** k * (1 - p) ** (n - k)
    return final
```

```python
#Gaussian dist, n = 100, p = 0.5 as a function of x
mu, sigma = 50, 0.5
g = np.random.normal(mu, sigma, 100)
plt.hist(g);
```

```python
#Binom dist, where mean and std and normalization same as binom, overplot
n, p = 100, 0.5
b = np.random.binomial(n, p, 100) #random binomial
plt.hist(b, color = 'r');


mu, sigma = 50, 5
g = np.random.normal(mu, sigma, 100) #random gaussian
plt.hist(g);

plt.show()
#note that binom dist is red, gauss dist is blue.
```

```python
#Binom dist, where mean and std and normalization same as binom, overplot
#changed from histogram as it masked a lot of data, although lines aren't
#much better anyways
n, p = 100, 0.5
b = np.random.binomial(n, p, 100) #random binomial
plt.plot(b, color = 'r');


mu, sigma = 50, 5
g = np.random.normal(mu, sigma, 100) #random gaussian
plt.plot(g);

plt.show()
#note that binom dist is red, gauss dist is blue.
```

# Q6

```python
n, p = 100, 0.02
b = np.random.binomial(n, p, 100) #random binom #2
plt.hist(b, color = 'r')
plt.xlim(-5, 20)

mu, sigma = 3, 3
g = np.random.normal(mu, sigma, 100) #random gaussian #2
plt.hist(g)

plt.show()
```

```python
n, p = 100, 0.02
b = np.random.binomial(n, p, 100) #random binom #2
plt.plot(b, color = 'r')
plt.xlim(-5, 20)

mu, sigma = 3, 3
g = np.random.normal(mu, sigma, 100) #random gaussian #2
plt.plot(g)

plt.show()
```

Looks like our distributions match up relatively well here, although it doesn't look that way in the histogram which is odd.  We expect that at high N the shapes of our distributions will be quite similar.  The difference between the two is that a binomial distribution is discrete whereas our normal distribution is continuous.  This makes it hard to plot since it's not possible to find data values between our calculated points.
