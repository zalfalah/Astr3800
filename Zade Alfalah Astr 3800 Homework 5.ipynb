```python
from numpy import sqrt
import numpy as np
import matplotlib.pyplot as plt
```

```python
def s_n3(tt = 3600, x = 10, y = 2):#tt is total time, x is target counts per second, y is sky counts per second
    t_arr = np.linspace(0, tt, tt) #gives our x axis of times
    S_N = [] #empty values of sn to append to
    for t1 in t_arr: #loop over all times on target
        t2 = tt - t1 #defines time on sky
        R = t1/t2 #ratio of time on target vs sky
        S_N.append(((x+y)*t1-R*y*t2)/np.sqrt((R*np.sqrt(y*t2))**2+(x+y)*t1)) #formula to calculate sn with variables above
    plt.plot(S_N) #plots sn vs time on target
    plt.xlabel("Time on Target (sec)")
    plt.ylabel("S/N Value")
    plt.title("S/N Vs. Time")
    S_arr = np.asarray(S_N) #makes our sn values an array
    S_arr = S_arr[~np.isnan(S_arr)] #gets rid of nan values from division by zero
    re = np.where(S_arr==S_arr.max())
    #print(S_arr.max(), np.where(S_arr==S_arr.max()))#give the best sn possible and the time on target
    print(f"The best S/N we can get is {S_arr.max()} at a time of {re} (sec) on target") #prints best sn, and time on target
```

```python
s_n3()
```

```python
s_n3(y = 5)
```
